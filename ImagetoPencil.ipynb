Importing Required Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
from google.colab.patches import cv2_imshow
```

Read the Image

```python
img = cv2.imread('virat.jpg', cv2.IMREAD_UNCHANGED)
plt.imshow(img)
```

Converting the original image to greyscale:

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image)
```

Invert the new grayscale image:

```python
inverted_image = cv2.bitwise_not(gray_image)
plt.imshow(inverted_image)
```

Blur the image

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
```

Generating the pencil sketch

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
display(Image.fromarray(pencil_sketch))
```

Comparing with the old vs new image

```python
display(Image.fromarray(img))
display(Image.fromarray(pencil_sketch))
```
